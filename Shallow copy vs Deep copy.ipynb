# Assignment vs Shallow vs Deepcopy

# Assignment

```python
lst1 = [1,2,3,4]
```

```python
lst2 = lst1
```

```python
lst1
```

```python
lst2
```

```python
lst2[1] = 10
```

```python
lst2
```

```python
lst1
```

# Shallow copy

```python
lst1 = [1,2,3,4]
```

```python
lst2 = lst1.copy()
```

```python
lst1
```

```python
lst2
```

```python
lst2[1] = 10
```

```python
lst2
```

```python
lst1
```

```python
lst1 = [[1,2,3,4],[5,6,7,8]]
```

```python
lst2 = lst1.copy()
```

```python
lst1
```

```python
lst2
```

```python
lst2[1][0] = 10
```

```python
lst2
```

```python
lst1
```

```python
lst2[0].append(5)
```

```python
lst2
```

```python
lst1
```

# Deep copy

```python
lst1 = [1,2,3,4]
```

```python
import copy
```

```python
lst2 = copy.deepcopy(lst1)
```
